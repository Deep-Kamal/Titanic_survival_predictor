```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

# Data cleaning

```python
df.drop(columns=['zero','zero.1','zero.2','zero.3','zero.4','zero.5','zero.6','zero.7','zero.8','zero.9','zero.10','zero.11','zero.12','zero.13','zero.14','zero.15','zero.16','zero.17','zero.18'],inplace=True)
```

```python
df
```

```python
df.drop(columns=['Pclass'],inplace=True)
```

```python
df
```

```python
# step 1 -> train/test/split
X_train, X_test, Y_train, Y_test = train_test_split(df.drop(columns=['2urvived']),
                                                   df['2urvived'],
                                                   test_size=0.2,
                                                   random_state=42)
```

```python
X_train.head()
```

```python
Y_train.sample(6)
```

```python
# imputation Transformer
trf1 = ColumnTransformer([
    ('impute_age', SimpleImputer(),[1]),
    ('imput_Embarked',SimpleImputer(strategy='most_frequent'),[6])  
],remainder='passthrough')
```

```python
# one Hot encodeing 
trf2 = ColumnTransformer([
    ('ohe_sex_embarked', OneHotEncoder(sparse_output= False, handle_unknown='ignore'), [1, 6])
],remainder='passthrough')
```

```python
#scaling 
trf3 = ColumnTransformer([
    ('scale', MinMaxScaler(), slice(0, 10))
])
```

```python
# Feacture Selection 
trf4 = SelectKBest(score_func=chi2, k = 5)
```

```python
# train the model 
trf5 = DecisionTreeClassifier() 
```

# Create Pipeline

```python
Pipe = Pipeline([
    ('trf1', trf1), 
    ('trf2', trf2), 
    ('trf3', trf3), 
    ('trf4', trf4), 
    ('trf5', trf5)
])
```

```python
# train 
Pipe.fit(X_train,Y_train)
```

# ML Model 

```python
# predict 
y_pred = Pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, y_pred)
```
